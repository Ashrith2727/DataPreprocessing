```python
import pandas as pd
import numpy as np
df = pd.read_csv('cars.csv')

#Display the first few rows of the dataset
#print(df.head())
df.describe()
```

```python
#Drop rows with missing values
df = df.dropna()
df1 = df.reset_index(drop=True)
df1
```

```python
#Identify duplicates
print(df1.duplicated().sum())

#Remove duplicates
#df_no_duplicates = df.drop_duplicates()
```

```python
df1.isnull().sum()
```

```python
df1.to_csv('final.csv')
df2 = pd.read_csv('final.csv', index_col = 0)
df2
```

```python
df3 = df2.drop(['model','interior','seller','saledate','transmission','trim', 'vin', 'color', 'body'], axis = 'columns')
dummies = pd.get_dummies(df3.make).astype(np.int8)
dummies.drop(['Volvo', 'Acura', 'Audi', 'BMW', 'Lamborghini'], axis = 'columns')
#dummies_final
dummies_update = dummies[['Volvo', 'Acura', 'Audi', 'BMW', 'Lamborghini']]

#merging the dummies and variable
merged = pd.concat([df3, dummies_update], axis = 'columns')
merged
```

```python
final = merged.drop(['make'], axis = 'columns')
final = final.drop_duplicates()
print(final.duplicated().sum())
```

```python
final.to_csv('final_test.csv')
finale = pd.read_csv('final_test.csv', index_col = 0)
finale = finale.drop_duplicates()
finale
```

```python
print(finale.duplicated().sum())
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

finale = pd.read_csv('final_test.csv', index_col = 0)
x = finale.drop(columns = ['sellingprice'])
y = finale['sellingprice']

x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.2)

model = DecisionTreeClassifier()

model.fit(x_train, y_train)
predictions = model.predict(x_test)

model.predict([[2014, 100000, 40000, 0, 1, 0, 0, 0]])
```
